```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:\\Users\\RONKE\\Downloads\\output.csv")
df
```

```python
df.sample(50)
```

```python
#sorting df by year to view recent year 
df = df.sort_values(by=['Year','State'],ascending=False).reset_index(drop=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.describe(include="all")
```

```python
df.info()
```

```python
sns.set(rc={'figure.figsize':(15,7)})
ax=sns.barplot(data=df,x='State',y='Rate')
for item in ax.get_xticklabels():#to give room for the x axis labels
    item.set_rotation(90)
    
ax.set_title("Unemployment Rate by State")
plt.savefig('unemployment_rate_by_state.png')
```

```python
sns.set(rc={'figure.figsize':(15,7)})
ax=sns.barplot(data=df,x='Year',y='Rate')
for item in ax.get_xticklabels():#to give room for the x axis labels
    item.set_rotation(90)
    
ax.set_title("Unemployment Rate by Year")
plt.savefig('unemployment_rate_by_Year.png')
```

This statistic displays the unemployment rate in the U.S. from 1990 to 2016. In 1990, this rate stood at 6.0 percent. Unemployment rate stood at 5.1 percent in 2016.

```python
# filter rows for year 2010 using  the boolean expression
year_2010 = df[df.Year.eq(2010)]
year_2010
```

```python
#checking for state with highest unemployment rate in 2010
sns.set(rc={'figure.figsize':(15,7)})
ax=sns.barplot(data=year_2010,x='State',y='Rate')
for item in ax.get_xticklabels():#to give room for the x axis labels
    item.set_rotation(90)
    
ax.set_title("Unemployment Rate by State for year 2010")
plt.savefig('unemployment_rate_by_State_for_year_2010.png')
```
