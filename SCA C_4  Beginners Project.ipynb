#### Question 2
Guess The Number: Write a program where the computer randomly generates a number between 0 and 20. The user needs to guess what the number is. If the user guesses wrong, tell them their guess is either too high, or too low.

```python
import random
number=random.randint(0,20) 
guess = False #This can be anything bool/float/etc.. as long it is not an int
while type(guess) is not int:
    try:
        guess =input('Enter a number : ') #Cast input 
    
        if int(guess) < number:
            print('your guess is too low')
        if int(guess) > number:
            print('your guess is too high')
        if int(guess) == number:
            print(number,'you guessed right')
            break
    except ValueError:   #Handles all non int till the an integer value is entered
        print('Not correct format.Guess must be an integer')
        
```
